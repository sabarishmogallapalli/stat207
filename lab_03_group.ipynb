How does the association between Flavor and Aroma change based on different values for Country in this dataset?

direction, shape, strength, slope, outliers

````python
sns.lmplot(x="Data.Scores.Flavor",y="Data.Scores.Aroma",hue="Location.Country", ci=False, data=df)
plt.show()
````

````python
sns.lmplot(x="Data.Scores.Flavor",y="Data.Scores.Aroma",hue="Location.Country", ci=False, data=df)
plt.xlim(6, 10)
plt.ylim(6, 10)
plt.figure(figsize=(10, 6))
plt.show()
````

````python
df[['Data.Scores.Flavor','Data.Scores.Aroma','Location.Country']].groupby('Location.Country').corr()
````

direction: Most all of the relationships between aroma and flavor in all countries is positive. There are two major exceptions: Nicagaragua and Zambia.   
shape: All of the relationships are linear.
strength: The aroma and flavor seem to have a strong relationship, all of their correlations are close to 1, ranging from 0.7-1. This shows a strong association between both aroma and flavor and this is consistent over most all of the countries.
slope: Brazil seems to be the strongest and most consistent slope. However almost everyone had a similar, positive slope. 
outliers: There seems to be one major outlier which is from Columbia.
