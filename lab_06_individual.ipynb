# <u>Case Study 1</u>: Predicting Coffee Flavor *for New Coffee Brands*



### Coffee Dataset Information

In this individual assignment we will return to the `coffee.csv` dataset. Information about the variables in this dataset can be found here: https://corgis-edu.github.io/corgis/csv/coffee/


### Research Goals

In this individual assignment we would ideally like to pursue multiple research goals.

#### Main Research Goal

Our main research goal will be to create a predictive model that effectively predicts the `Flavor` of *new coffee brands*.


#### Secondary Research Goals

Ideally, the predictive model that we select will also be able to accurately reflect the relationship between the explanatory variables and the response variable `Flavor`.






```python
import pandas as pd
import seaborn as sns
sns.set()
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

## 1. Coffee Dataset

### 1.1. Reading the csv

1. Read the `coffee.csv` into a dataframe. The strings that are used to represent missing values in this csv file are represented with 'nan'. Luckily, 'nan' is one of the string values that the **pd.read_csv()** function is automatically programmed to detect as a missing value and thus convert into a NaN type object.
2. Show the first 5 rows.
3. Show how many rows this dataframe has.

```python
coffee_df = pd.read_csv('coffee.csv')
coffee_df.head()
```

```python
print(f"The dataset has {coffee_df.shape[0]} rows.")
```

### 1.2. Dropping NaN Values

We only intend to use the `Flavor`, `Aroma`, `Aftertaste`, `Color`, and `Species` variables in this dataframe.

1. Create a dataframe with just these variables.
2. Then drop all rows from this dataframe that have a NaN value.
3. How many rows did you drop?

```python
coffee_five_df = coffee_df.copy()
coffee_five_df = coffee_five_df[["Data_Scores_Flavor","Data_Scores_Aroma","Data_Scores_Aftertaste","Data_Color","Data_Type_Species"]]
coffee_five_df
```

```python
coffee_five_dropped_df = coffee_five_df.dropna()
coffee_five_dropped_df
```

```python
print(f"We dropped {coffee_five_df.shape[0] - coffee_five_dropped_df.shape[0]} rows")
```

### 1.3. Dropping a Large Outlier

This dataset has a prominent outlier with an `Aroma` rating that is equal to 0. Drop this observation from the dataset.

```python
aroma_zero = coffee_five_dropped_df.loc[coffee_five_dropped_df['Data_Scores_Aroma'] == 0]
aroma_zero
```

```python
coffee_five_dropped_df.drop(985)
```

## 2. Training and Test Datasets

We intend to build linear regression models that will predict the `Flavor` of *new coffee brands* using some of the following explanatory variables.

* `Aroma`
* `Aftertaste`
* `Color`
* `Species`


Using some combination of these explanatory variables, we'd like to build and select a linear regression model which we can infer might perform well when it comes to predicting the `Flavor` of *new coffee brands*. Thus, let's create a training and a test dataset from our cleaned dataframe from 1.3.

Use a random state of `437` to do this. Your training dataset should be 80% of observations from your cleaned dataframe.

```python
coffee_test_df,coffee_train_df = train_test_split(coffee_five_dropped_df,test_size = 0.2, random_state = 437)
print(coffee_test_df.shape[0] / coffee_five_dropped_df.shape[0])
```

```python
print(coffee_train_df.shape[0] / coffee_five_dropped_df.shape[0])
```

## 3. Full Model

### 3.1. Fitting the Full Model

Next, fit the "full" linear regression model that predicts `Flavor` with

* `Aroma`
* `Aftertaste`
* `Color`
* `Species`.

Display the summary output table for this model.

```python
full_model = smf.ols(formula='Data_Scores_Flavor~Data_Scores_Aroma+Data_Scores_Aftertaste+Data_Color+Data_Type_Species',data=coffee_train_df).fit()
full_model.summary()
```

### 3.2. Linear Regression Equation

Write out the linear regression equation for this model. Make sure to use appropriate notation.

Data_Scores_Flavor = 0.9764 + (Data_Color[T.Bluish-Green])(0.0350)+(Data_Color[T.Green])(0.0763)+(Data_Color[T.Unknown])(0.0482)+(Data_Type_Species[T.Robusta])(-0.0149)+(Data_Scores_Aroma)(0.2536)+(Data_Scores_Aftertaste)(0.6194)

### 3.3. Test Dataset Fit

Calculate the test R^2 of this full model.

```python
full_model_test_predict = full_model.predict(coffee_test_df)
```

```python
r2_score(coffee_test_df["Data_Scores_Flavor"],full_model_test_predict)
```

### 4.2. Incorporating Interaction Terms

Select the top three `categorical explanatory variable` and `numerical explanatory variable` pairs that you identified in 4.1 that display the *strongest* interaction between how they impact the `Flavor` response variable. Then incorporate the interaction terms that model this interaction effect of your three chosen `categorical explanatory variable` and `numerical explanatory variable` pairs into the full model.

Fit this new model with the **training dataset** again and display the summary output table.

    Species - Aroma
    Color - Aroma
    Species - Aftertaste

```python
full_model_three = smf.ols(formula='Data_Scores_Flavor~(Data_Type_Species * Data_Scores_Aroma)+(Data_Scores_Aroma * Data_Color)+(Data_Scores_Aftertaste * Data_Type_Species) + Data_Scores_Aftertaste + Data_Color + Data_Type_Species + Data_Scores_Aroma',data=coffee_train_df).fit()
full_model_three.summary()
```

### 4.3. Linear Regression Equation

Write out the linear regression equation for this model. Make sure to use appropriate notation.

Data_Scores_Flavor = 0.6507 + 3.1199(Data_Type_Species[T.Robusta])-0.8622(Data_Color[T.Bluish-Green])+0.4401(Data_Color[T.Green])+1.2900(Data_Color[T.Unknown])+0.2817(Data_Scores_Aroma)-0.3035(Data_Type_Species[T.Robusta] * Data_Scores_Aroma)+0.1179(Data_Scores_Aroma * Data_Color[T.Bluish-Green])-0.0473(Data_Scores_Aroma * Data_Color[T.Green])-0.1618(Data_Scores_Aroma * Data_Color[T.Unknown])+0.6336(Data_Scores_Aftertaste)-0.0981(Data_Scores_Aftertaste * Data_Type_Species[T.Robusta])

### 4.4. Test Dataset Fit

Calculate the test R^2 of this full model.

```python
z=full_model_three.predict(coffee_test_df)

r2_score(coffee_test_df["Data_Scores_Flavor"],z)
```
