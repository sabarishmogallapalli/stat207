### 1.3. Explaining the Result

How would you explain to a `STAT107 student` (who has never performed an **inner join merge** before) why the resulting merged dataframe in 1.2 resulted in 11 rows?

When we do an inner join merge with the first_name as the key, it takes the dataset we defined as the inner ("generation") and compares it to every row of the outer ("occupation") df. For each row, it checks if the first name matches up and if it does it creates a row merging the two datasets. That is why you see multiple of the tori kelly row from the occupation dataset, because it is being merged with each instance of tory that is found in the generation dataset. This is what results in 11 rows.
