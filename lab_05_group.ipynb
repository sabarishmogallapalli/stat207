## 2. <u>Research Question</u>: Most Independent Predictive Power

We would first like to answer the following research question using the *full dataset*.

`Which explanatory variable in this list brings the most amount of predictive power to the model by itself?`

### 2.1. 10 Candidate Models

Create 10 simple linear regression candidate models.

* One that predicts `Flavor` with `Aroma`
* One that predicts `Flavor` with `Aftertaste`
* One that predicts `Flavor` with `Acidity`
* One that predicts `Flavor` with`Body`
* One that predicts `Flavor` with `Balance`
* One that predicts `Flavor` with `Uniformity`
* One that predicts `Flavor` with `Sweetness`
* One that predicts `Flavor` with `Moisture`
* One that predicts `Flavor` with `Color`
* One that predicts `Flavor` with `Species`

Calculate/show the R^2 for each of these models.

```python
#One that predicts Flavor with Aroma
lr_flavor_aroma = smf.ols(formula='Data_Scores_Flavor~Data_Scores_Aroma',data=filtered_drop).fit()
lr_flavor_aroma.summary()
```

```python
#One that predicts Flavor with Aftertaste
lr_flavor_aftertaste = smf.ols(formula='Data_Scores_Flavor~Data_Scores_Aftertaste',data=filtered_drop).fit()
lr_flavor_aftertaste.summary()
```

```python
#One that predicts Flavor with Acidity
lr_flavor_acidity = smf.ols(formula='Data_Scores_Flavor~Data_Scores_Acidity',data=filtered_drop).fit()
lr_flavor_acidity.summary()
```

```python
#One that predicts Flavor with Body
lr_flavor_body = smf.ols(formula="Data_Scores_Flavor~Data_Scores_Body",data=filtered_drop).fit()
lr_flavor_body.summary()
```

```python
#One that predicts Flavor with Balance
lr_flavor_balance = smf.ols(formula="Data_Scores_Flavor~Data_Scores_Balance",data=filtered_drop).fit()
lr_flavor_balance.summary()
```

```python
#One that predicts Flavor with Uniformity
lr_flavor_uniformity = smf.ols(formula="Data_Scores_Flavor~Data_Scores_Uniformity",data=filtered_drop).fit()
lr_flavor_uniformity.summary()
```

```python
#One that predicts Flavor with Sweetness
lr_flavor_sweetness = smf.ols(formula="Data_Scores_Flavor~Data_Scores_Sweetness",data=filtered_drop).fit()
lr_flavor_sweetness.summary()
```

```python
#One that predicts Flavor with Moisture
lr_flavor_moisture = smf.ols(formula="Data_Scores_Flavor~Data_Scores_Moisture",data=filtered_drop).fit()
lr_flavor_moisture.summary()
```

```python
#One that predicts Flavor with Color
lr_flavor_color = smf.ols(formula="Data_Scores_Flavor~Data_Color",data=filtered_drop).fit()
lr_flavor_color.summary()
```

```python
#One that predicts Flavor with Species
lr_flavor_species = smf.ols(formula="Data_Scores_Flavor~Data_Type_Species",data=filtered_drop).fit()
lr_flavor_species.summary()
```

### 2.2. Answer

Use your analyses from 2.1 to answer the research question.

Aftertaste had the highest r^2 value of 0.755 which means aftertaste has the highest predictive power on Flavor.
